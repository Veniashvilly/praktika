**Задание 2: Анализ архитектур CNN (30 баллов)**\
**2.1 Влияние размера ядра свертки (15 баллов)**

```python
import torch
from datasets import get_mnist_loaders
from models import NEWCNN
from trainer import train_model
from utils import show_first_layer_activations
import time
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

batch_size = 64
epochs = 5
lr = 0.001
num_classes = 10

train_loader, test_loader = get_mnist_loaders(batch_size=batch_size)

print('Ядро 3*3')
cnn_3x3 = NEWCNN(kernel_size=3).to(device)
start = time.time()
cnn_3x3_history = train_model(cnn_3x3, train_loader, test_loader, epochs=epochs, lr=lr, device=device)
end = time.time()
print(end-start)
show_first_layer_activations(cnn_3x3, test_loader, device)

print('Ядро 5*5')
cnn_5x5 = NEWCNN(kernel_size=5).to(device)
start = time.time()
cnn_5x5_history = train_model(cnn_5x5, train_loader, test_loader, epochs=epochs, lr=lr, device=device)
end = time.time()
print(end-start)
show_first_layer_activations(cnn_5x5, test_loader, device)

print('Ядро 7*7')
cnn_7x7 = NEWCNN(kernel_size=7).to(device)
start = time.time()
cnn_7x7_history = train_model(cnn_7x7, train_loader, test_loader, epochs=epochs, lr=lr, device=device)
end = time.time()
print(end-start)
show_first_layer_activations(cnn_7x7, test_loader, device)

print('COMBO')
cnn_combo = NEWCNN(kernel_size=5, combo=1).to(device)
start = time.time()
cnn_combo_history = train_model(cnn_combo, train_loader, test_loader, epochs=epochs, lr=lr, device=device)
end = time.time()
print(end-start)
show_first_layer_activations(cnn_combo, test_loader, device)
```

**2.2 Влияние глубины CNN**

```python
import torch
from datasets import get_mnist_loaders
from models import CNN_4layers,CNN_6layers,SimpleCNN,CNNWithResidual
from trainer import train_model
from utils import feature_maps
import time
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

batch_size = 64
epochs = 5
lr = 0.001
num_classes = 10

train_loader, test_loader = get_mnist_loaders(batch_size=batch_size)

print('Неглубокая CNN')
not_deep = SimpleCNN(input_channels=1, num_classes=num_classes).to(device)
start = time.time()
not_deep_history = train_model(not_deep, train_loader, test_loader, epochs=epochs, lr=lr, device=device, is_grad_flow=1)
end = time.time()
print(end-start)

print('Средняя CNN')
mid = CNN_4layers(input_channels=1, num_classes=num_classes).to(device)
start = time.time()
mid_history = train_model(mid, train_loader, test_loader, epochs=epochs, lr=lr, device=device, is_grad_flow=1)
end = time.time()
print(end-start)

print('Глубокая CNN')
deep = CNN_6layers(input_channels=1, num_classes=num_classes).to(device)
start = time.time()
deep_history = train_model(deep, train_loader, test_loader, epochs=epochs, lr=lr, device=device, is_grad_flow=1)
end = time.time()
print(end-start)

print('CNN с Residual связями')
cnn_with_red = CNNWithResidual(input_channels=1, num_classes=num_classes).to(device)
start = time.time()
cnn_with_red_history = train_model(cnn_with_red, train_loader, test_loader, epochs=epochs, lr=lr, device=device, is_grad_flow=1)
end = time.time()
print(end-start)

#Визуализация feature maps
feature_maps(not_deep, test_loader, device, model_name="SimpleCNN")
feature_maps(mid, test_loader, device, model_name="CNN_4layers")
feature_maps(deep, test_loader, device, model_name="CNN_6layers")
feature_maps(cnn_with_red, test_loader, device, model_name="CNNWithResidual")
```
